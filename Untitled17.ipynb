```python
import pandas as pd
opti_path = 'https://github.com/shimonjacobs/dataset/blob/main/optitrack_recording.csv?raw=true'
odom_path = 'https://github.com/shimonjacobs/dataset/blob/main/odometry.csv?raw=true'
optitrack_columns = ['Frame', 'Time (Seconds)', 'Quaternion_X', 'Quaternion_Y', 'Quaternion_Z', 'Quaternion_W', 'pos_x', 'pos_y', 'pos_z']
odometry_df = pd.read_csv(odom_path)
optitrack_df = pd.read_csv(opti_path,  skiprows=7, names=optitrack_columns)
```

```python
#Initial Data

import matplotlib.pyplot as plt

fig, axes = plt.subplots(6, 2, figsize=(12, 18))

# Odometry plots
axes[0, 0].plot(odometry_df['pos_x'])
axes[0, 0].set_title('Odometry X')
axes[1, 0].plot(odometry_df['pos_y'])
axes[1, 0].set_title('Odometry Y')
axes[2, 0].plot(odometry_df['pos_z'])
axes[2, 0].set_title('Odometry Z')

# OptiTrack plots
axes[0, 1].plot(optitrack_df['pos_x'])
axes[0, 1].set_title('OptiTrack X')
axes[1, 1].plot(optitrack_df['pos_y'])
axes[1, 1].set_title('OptiTrack Y')
axes[2, 1].plot(optitrack_df['pos_z'])
axes[2, 1].set_title('OptiTrack Z')

# Additional plots (empty for now)
axes[3, 0].set_visible(False)
axes[3, 1].set_visible(False)
axes[4, 0].set_visible(False)
axes[4, 1].set_visible(False)
axes[5, 0].set_visible(False)
axes[5, 1].set_visible(False)

plt.tight_layout()
plt.show()

```

```python
# Process Data


# convert optitrack to FLU
optitrack_df_flu = optitrack_df.copy()
optitrack_df_flu['pos_y'] = optitrack_df['pos_z']
optitrack_df_flu['pos_z'] = -optitrack_df['pos_y']

#align data
opti_len = optitrack_df['pos_x'].shape[0]
adjust = 2100

odom_final = odometry_df.iloc[adjust:adjust+opti_len]
opti_final = optitrack_df_flu.copy()

# # Shift data to make the starting point (0, 0, 0)
odom_final['pos_x'] -= odom_final['pos_x'].iloc[0]
odom_final['pos_y'] -= odom_final['pos_y'].iloc[0]
odom_final['pos_z'] -= odom_final['pos_z'].iloc[0]

opti_final['pos_x'] -= opti_final['pos_x'].iloc[0]
opti_final['pos_y'] -= opti_final['pos_y'].iloc[0]
opti_final['pos_z'] -= opti_final['pos_z'].iloc[0]
```

```python
# prompt: repeat cell 2 but with final variable

#Final Data


fig, axes = plt.subplots(6, 2, figsize=(12, 18))

# Odometry plots
axes[0, 0].plot(odom_final['pos_x'])
axes[0, 0].set_title('Odometry X')
axes[1, 0].plot(odom_final['pos_y'])
axes[1, 0].set_title('Odometry Y')
axes[2, 0].plot(odom_final['pos_z'])
axes[2, 0].set_title('Odometry Z')

# OptiTrack plots
axes[0, 1].plot(opti_final['pos_x'])
axes[0, 1].set_title('OptiTrack X')
axes[1, 1].plot(opti_final['pos_y'])
axes[1, 1].set_title('OptiTrack Y')
axes[2, 1].plot(opti_final['pos_z'])
axes[2, 1].set_title('OptiTrack Z')

# Additional plots (empty for now)
axes[3, 0].set_visible(False)
axes[3, 1].set_visible(False)
axes[4, 0].set_visible(False)
axes[4, 1].set_visible(False)
axes[5, 0].set_visible(False)
axes[5, 1].set_visible(False)

plt.tight_layout()
plt.show()
```
